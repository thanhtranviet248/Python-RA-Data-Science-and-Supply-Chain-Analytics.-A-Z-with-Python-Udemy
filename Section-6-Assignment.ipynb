```python
import pandas as pd
import numpy as np
```

```python
airport = pd.read_csv("airports.csv")
flight = pd.read_csv("flights.csv")
plane = pd.read_csv("planes.csv")
airline = pd.read_csv("airlines.csv")
A = (airport,flight,plane,airline)

for i in range (len(A)):
    print (A[i].head(5))
```

```python
#1/ what is the most popular destination city from NewYork?
popular_dest = flight[["dest"]].value_counts().sort_values(ascending=[0])
popular_dest
pd.merge(popular_dest,airport[["faa","name"]],how="left",left_on="dest",right_on="faa")
```

```python
#2/ which month is the busiest of the year?
busiest_month = flight[["year","month"]].value_counts().sort_values(ascending=[0])
busiest_month
```

```python
#3/ which airline is the most punctual?
#6/ what airline is the worst in terms of delays
flight_airline = pd.merge(flight,airline,how="left")
flight_airline = flight_airline[["dep_delay","arr_delay","name"]]
flight_airline
```

```python
for col in flight_airline.select_dtypes(include=['float']):
    flight_airline[col] = flight_airline[col].apply(lambda x: abs(x) if x < 0 else x)
flight_airline
```

```python
flight_airline = pd.pivot_table(flight_airline,columns="name",values=["dep_delay","arr_delay"],fill_value=0,aggfunc="mean")
flight_airline
```

```python
max_delay = flight_airline.idxmax(axis=1)
print(max_delay)
min_delay = flight_airline.idxmin(axis=1)
print(min_delay)
```

```python
#4/ what destination is the longest duration
flight.head()
```

```python
longest_duration = flight[["air_time","distance","dest"]]
longest_duration = pd.pivot_table(longest_duration,columns="dest",values=["air_time","distance"],fill_value=0,aggfunc="mean")
longest_duration
```

```python
longest_duration = longest_duration.idxmax(axis=1)
longest_duration
```

```python
#6/ which airline has the highest capacity of seats?
print(airline.head())
print(flight.head())
print(plane.head())
```

```python
airline_flight = pd.merge(airline,flight,how="left")
airline_flight.head()
```

```python
airline_flight_plane = pd.merge(airline_flight,plane,on="tailnum",how="left")
airline_flight_plane.head()
```

```python
highest_cap_seats = airline_flight_plane[["name","tailnum","seats"]]
highest_cap_seats.head()
```

```python
highest_cap_seats.value_counts().sort_values(ascending=[0])
```

```python
#7/ which airplane model is the highest in use and from which manufacturer?
flight_plane = pd.merge(flight,plane,on="tailnum",how="left")
flight_plane_pivot = pd.pivot_table(flight_plane,columns=["manufacturer","model"],values="air_time",aggfunc="count")
flight_plane_pivot
```

```python
flight_plane_pivot.idxmax(axis=1)
```
