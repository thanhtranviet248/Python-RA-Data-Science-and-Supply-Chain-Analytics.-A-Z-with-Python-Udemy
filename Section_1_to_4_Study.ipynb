```python
#conditions

def status(person):
  if person >= 20: print("adult")
  elif person >= 14: print("teenage")
  else : print("kids")

A = 10
B = 15
C = 50

class_list = [A, B, C]

map_object = map(status,class_list)
status_list = list(map_object)
```

```python
for i in class_list:
  print(i)
```

```python
def status(person):
  if person >= 20: return("adult")
  elif person >= 14: return("teenage")
  else : return("kids")

new_status_list = []
for age in class_list:
  new_status_list.append(status(age))
print(new_status_list)
```

```python
retail_data = pd.read_csv("online_retail2.csv")
retail_data
```

```python
def check_UK(country):
  if country == "United Kingdom": return "Yes"
  else: return "No"

check_UK("United Kingdom")
```

```python
retail_data["UK"] = retail_data["Country"].map(check_UK)
retail_data["UK"]

retail_data["UK1"] = np.NaN

first_10 = retail_data.iloc[:10,:]
first_10_shape = first_10.shape[0]

for i in range(0,first_10_shape):
  first_10.loc[i,"UK1"] = check_UK(first_10.Country[i])

first_10
```
