```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("cars.csv")
data.head()
```

```python
data.shape
```

```python
unique_cylenders = data["cylenders"].unique()
unique_cylenders
```

```python
data.describe()
```

```python
highest_price = data['Price'].max()
highest_price

most_expensive_car = data[data['Price'] == highest_price]
most_expensive_car = most_expensive_car['name']
most_expensive_car
```

```python
data_rename = data.rename(columns={'name':'car name'})
data_rename
```

```python
car_pricing = data_rename.loc[:,['car name','Price']]
car_pricing["Price"] = car_pricing["Price"].astype(int)

def pricing_cat(Price):
  if Price >= 35000: return "Expensive Car"
  elif Price >= 20000: return "Suitable Car"
  else: return "Budget Car"

car_pricing["Category"]=np.NaN

for i in range(0,car_pricing.shape[0]):
  car_pricing.loc[i,"Category"]=pricing_cat(car_pricing["Price"][i])

car_pricing
```

```python
budget_car = car_pricing[car_pricing["Category"] == "Budget Car"].shape[0]
suitable_car = car_pricing[car_pricing["Category"] == "Suitable Car"].shape[0]
expensive_car = car_pricing[car_pricing["Category"] == "Expensive Car"].shape[0]
print("Budget Car:", budget_car)
print("Suitable Car:", suitable_car)
print("Expensive Car:", expensive_car)
```
