```python
## joining

designation = pd.DataFrame({"Name":["Mike","Jonathan","Mo","Lisa","Raj","Teo"],
                            "Title":["Manager","Director","Supervisor","Associate","Assistant","Sectionhead"]})

age = pd.DataFrame({"Name":["Mike","Jonathan","Lee","Lisa","Tom","Teo"],
                            "Age":["40","50","34","35","24","60"]})

print(designation)
print(age)
```

```python
#left join
left_join = pd.merge(designation,age,how="left")
left_join
```

```python
#right join
right_join = pd.merge(designation,age,how="right")
right_join
```

```python
#outer join
full_join = pd.merge(designation,age,how="outer")
full_join
```

```python
#ineer join
inner_join = pd.merge(designation,age,how="inner")
inner_join
```

```python
inner_join[inner_join["Name"]=="Mike"]
```
